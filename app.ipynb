```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Extrair

df = pd.read_csv('winequality-red.csv')

df.head()
```

```python
df.describe()
```

```python
print(df.isnull().sum())
```

### Gráfico de correlação

```python
# Correlação
    
correlacao = df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlacao, annot=True, fmt='.1f', cmap='crest')
plt.show()

# A qualidade do vinho depende muito do alcohol, sulphates e o citric acid
```

### Relação do alcool com a Sulfatos


```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x="alcohol", y="sulphates", hue='quality', data=df, alpha=0.8)
plt.title("Relação entre Alcool com Sulfatos")
plt.xlabel("Teor Alcoólico")
plt.ylabel("Sulfatos")
plt.xticks(rotation=45)
plt.show()
```

### Distribuição do Teor Alcoólico

```python
plt.figure(figsize=(8, 6))
sns.histplot(df["alcohol"], bins=35, kde=True, color="red")
plt.title("Distribuição do Teor Alcoólico no Vinho")
plt.xlabel("Teor Alcoólico (%)")
plt.ylabel("Frequência")
plt.show()
```
